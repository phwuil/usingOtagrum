```python
import otagrum

import openturns as ot

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb
```

```python
R = ot.CorrelationMatrix(2)
R[0,1] = 0.8
ncopula = ot.Normal([0.0]*2, [1.0]*2, R).getCopula()

X = Y = np.linspace(0, 1, num=20)
X,Y = np.meshgrid(X,Y)

Z = list(zip(X.flatten(),Y.flatten()))

Z = [ncopula.computePDF(ot.Point(z)) for z in Z]

Z = np.array(Z).reshape(X.shape)

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
```

```python
d = 4
R = ot.CorrelationMatrix(d)
for i in range(R.getDimension()):
    for j in range(i):
        R[i,j] = 0.8
```

```python
normal_copula = ot.Normal([0.0]*d, [1.0]*d, R).getCopula()
```

```python
point = ot.Point([0.1,0.2,0.5,0.7])
```

```python
normal_copula_margin = normal_copula.getMarginal(list(range(1,d)))
```

```python
normal_copula.computePDF(point)/normal_copula_margin.computePDF(point[1:d])
```

```python
normal_copula.computeConditionalPDF(point[0], point[1:d])
```
