```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('text', usetex=True)  
mpl.rc('font', family='serif')
import sys
sys.path.append('..')
import utils as ut
```

```python
lw = 2.5
label_fs = 20
tick_fs = 16
legend_fs = 16
small_legend_fs = 14
```

```python
cpc_gaussian = np.loadtxt('../../results/gaussian/asia/r08/scores/skeleton_scores_cpc_asia_gaussian_f50t5000s10r5mcss5alpha5.csv',delimiter=',').transpose()
cpc_student = np.loadtxt('../../results/student/asia/r08/scores/skeleton_scores_cpc_asia_student_f50t5000s10r5mcss5alpha5.csv', delimiter=',').transpose()
cpc_dirichlet = np.loadtxt('../../results/dirichlet/asia/scores/skeleton_scores_cpc_asia_dirichlet_f50t5000s10r5mcss5alpha5.csv', delimiter=',').transpose()
elidan_gaussian = np.loadtxt('../../results/gaussian/asia/r08/scores/skeleton_scores_elidan_asia_gaussian_f50t5000s10r5mp4hcr3.csv', delimiter=',').transpose()
elidan_student = np.loadtxt('../../results/student/asia/r08/scores/skeleton_scores_elidan_asia_student_f50t5000s10r5mp4hcr3.csv', delimiter=',').transpose()
elidan_dirichlet = np.loadtxt('../../results/dirichlet/asia/scores/skeleton_scores_elidan_asia_dirichlet_f50t5000s10r5mp4hcr3.csv', delimiter=',').transpose()
```

```python
sizes = cpc_gaussian[0].astype(int)
```

```python
mean_fscore_cpc_gaussian, std_fscore_cpc_gaussian = cpc_gaussian[3], cpc_gaussian[6]
mean_fscore_cpc_student, std_fscore_cpc_student = cpc_student[3], cpc_student[6]
mean_fscore_cpc_dirichlet, std_fscore_cpc_dirichlet = cpc_dirichlet[3], cpc_dirichlet[6]

mean_fscore_elidan_gaussian, std_fscore_elidan_gaussian = elidan_gaussian[3], elidan_gaussian[6]
mean_fscore_elidan_student, std_fscore_elidan_student = elidan_student[3], elidan_student[6]
mean_fscore_elidan_dirichlet, std_fscore_elidan_dirichlet = elidan_dirichlet[3], elidan_dirichlet[6]
```

```python
fig, ax = plt.subplots()

ax.set_xlabel('Size', fontsize=label_fs)
ax.set_ylabel('F-score', fontsize=label_fs)

ax.tick_params(labelsize=tick_fs)

ax.set_xlim([sizes[0], sizes[-1]])
ax.set_ylim(0,1)

ax.plot(sizes, mean_fscore_cpc_gaussian, linestyle='--', linewidth=lw, label='gaussian')
#ut.plot_error(sizes, mean_fscore_cpc_gaussian, std_fscore_cpc_gaussian)

ax.plot(sizes, mean_fscore_cpc_student, linestyle='-.', linewidth=lw, label='student')
#ut.plot_error(sizes, mean_fscore_cpc_student, std_fscore_cpc_student)

ax.plot(sizes, mean_fscore_cpc_dirichlet, linestyle=':', linewidth=lw, label='dirichlet')
#ut.plot_error(sizes, mean_fscore_cpc_dirichlet, std_fscore_cpc_dirichlet)

ax.legend(loc='lower right', fontsize=legend_fs)
plt.savefig('fscore_skeleton_cpc.pdf', transparent=True)
plt.show()
```
